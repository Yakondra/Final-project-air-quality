```python
# Import of libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3 as sql
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
```

```python
pollution_2020 = pd.read_csv('../data/raw/pollution data/2020.csv', sep = ';')
pollution_2020.head()
```

```python
weather_2020 = pd.read_csv('../data/raw/weather data/2020.csv', sep = ';')
weather_2020.head()
```

```python
air_quality = pd.concat([weather_2020, pollution_2020], axis=0)
air_quality
```

```python
# Eliminar las columnas que no son necesarias
air_quality.drop(["provincia", "punto_muestreo"], axis=1, inplace=True)

# Eliminar las columnas de validacion que contengan "v"
air_quality = air_quality.drop(columns=[col for col in air_quality.columns if 'v' in col])

# Crear una columna 'fecha' con el ao, mes y da
air_quality['fecha'] = pd.to_datetime(air_quality[['ano', 'mes', 'dia']].astype(str).agg('-'.join, axis=1) + ' 00:00', errors='coerce', format='%Y-%m-%d %H:%M')
air_quality.drop(["ano", "mes", "dia"], axis=1, inplace=True)
```
